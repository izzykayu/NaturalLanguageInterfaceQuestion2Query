```python
import matplotlib.pyplot as plt
```

```python
with open("loss/loss_crf_noemb.txt", "r") as f:
    content = f.read()
f.close()
loss = content.split("\n")
loss = [float(l[2:-2]) for l in loss]
plt.plot(loss)
plt.title("bilstm crf without embedding")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.show()
```
