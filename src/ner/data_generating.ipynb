```python
import glob
import os
import re
import pickle
```

```python
txt_files = glob.glob("train/*.txt")
con_files = glob.glob("train/*.con")
files = [(txt_files[i], con_files[i]) for i in range(len(con_files))]
files
```

```python
training_data = []
list_fileSentence = []
list_fileEntity = []
list_fileLable = []
for txt_file, con_file in files:
    with open(txt_file) as tf:
        content = tf.read()
        lines = content.split("\n")
        sentence = [line.split() for line in lines]
        label = [["O" for w in list_word] for list_word in sentence] 
        list_fileSentence.append(sentence)
        list_fileLable.append(label)
    tf.close()

    list_groups = []
    with open(con_file) as cf:
        for line in cf.readlines():
            if not line.strip():
                continue
            concept_regex = '^c="(.*)" (\d+):(\d+) (\d+):(\d+)\|\|t="(\w*)"'
            match = re.search(concept_regex, line.strip())
            if match:
                groups = match.groups()
                list_groups.append([int(groups[1]) - 1, int(groups[2]), int(groups[4]), groups[5]])
        list_fileEntity.append(list_groups)
    cf.close()
    
for idx in range(len(list_fileSentence)):
    for entity in list_fileEntity[idx]:
        idx_line = entity[0]
        start = entity[1]
        end = entity[2]
        label = entity[3]
        #print(idx, idx_line, start, label)
        list_fileLable[idx][idx_line][start] = "B-" + label
        if start != end:
            for j in range(start+1, end+1):
                list_fileLable[idx][idx_line][j] = "I-" + label
    for idx_line in range(len(list_fileSentence[idx])):
        training_data.append((list_fileSentence[idx][idx_line], list_fileLable[idx][idx_line]))

len(training_data)
```
